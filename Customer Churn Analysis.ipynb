```python
data = df.groupby(['SeniorCitizen', 'Churn']).size().reset_index(name='Count')
data['Percentage'] = data['Count'] / data.groupby('SeniorCitizen')['Count'].transform('sum') * 100

# Prepare data for plotting
stacked_data = data.pivot(index='SeniorCitizen', columns='Churn', values='Percentage').fillna(0)

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(6, 4))
stacked_data.plot(kind='bar', stacked=True, ax=ax, colormap='viridis', alpha=0.85)

# Add labels
for i, row in enumerate(stacked_data.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:  # Only label non-zero values
            ax.text(
                x=i, y=cumulative + val / 2, s=f'{val:.1f}%', ha='center', va='center', fontsize=9, color='white'
            )
            cumulative += val

# Customize plot
plt.title('Churn by SeniorCitizen (Stacked Bar Chart (%))')
plt.xlabel('Senior Citizen')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)
plt.legend(title='Churn', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Define the columns to plot
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # Adjust rows/columns based on the number of features
axes = axes.flatten()  # Flatten the 2D array of axes for easier indexing

# Generate a countplot for each column
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i], palette="viridis", hue =df["Churn"])
    axes[i].set_title(f'Countplot of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Remove any empty subplots if the number of columns is less than the grid size
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout to avoid overlapping
plt.tight_layout()
plt.show()
```

#
This visualization illustrates the relationship between customer churn and their use of various services. Most customers have not churned (blue bars dominate), but certain categories, such as "Fiber optic" Internet and "No" for OnlineSecurity, OnlineBackup, and TechSupport, show higher churn rates (green bars). These patterns suggest that enhancing these specific services or addressing customer pain points could significantly reduce churn.
